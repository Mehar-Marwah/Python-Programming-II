```python
class RacingCar:
    def __init__(self):
        self.speed = 0   # initial speed is 0

    def acceleration(self, total_speed, total_time):
        # acceleration = total speed / total time
        acceleration = total_speed / total_time
        self.speed = total_speed
        return acceleration

    def braking(self, reduction):
        # reduce speed while braking
        self.speed = self.speed - reduction
        if self.speed < 0:
            self.speed = 0
        return self.speed


# Example usage
car = RacingCar()

acc = car.acceleration(100, 5)   # total speed = 100, total time = 5
print("Acceleration:", acc)

current_speed = car.braking(30)
print("Speed after braking:", current_speed)
```

```python
class Diet:
    def __init__(self):
        self.calories = 0   # initial calories

    def junk_calories(self, calories_added):
        self.calories = self.calories + calories_added
        return self.calories

    def exercise(self, calories_burned):
        self.calories = self.calories - calories_burned
        if self.calories < 0:
            self.calories = 0
        return self.calories

    def health_status(self):
        if self.calories > 30:
            return "Unhealthy"
        else:
            return "Healthy"


# Example usage
diet = Diet()

diet.junk_calories(40)
print("Calories:", diet.calories)
print("Health Status:", diet.health_status())

diet.exercise(15)
print("Calories after exercise:", diet.calories)
print("Health Status:", diet.health_status())
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # private attribute

    # method to get balance
    def get_balance(self):
        return self.__balance

    # deposit rule
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print("Amount deposited")
        else:
            print("Invalid deposit")

    # withdraw rule
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("Withdrawal successful")
        else:
            print("Insufficient balance")


# object creation
acc = BankAccount(1000)

acc.deposit(500)        # balance = 1500
acc.withdraw(200)       # balance = 1300
print(acc.get_balance())

# acc.__balance = -9999   # ❌ Not allowed (private variable)
# acc.withdraw(5000)      # ❌ Rule blocks it
```

```python
class IMBD:
    def __init__(self):
        self.rating = 0   # initial rating

    def new_rating(self, rating):
        self.rating = rating
        return self.rating


# object creation
movie = IMBD()

movie.new_rating(8.5)
print("Movie Rating:", movie.rating)
```
