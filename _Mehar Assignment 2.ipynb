```python
#Q1
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def car_info(self):
        return f"{self.year} {self.brand} {self.model}"


class ElectricCar(Car):
    def __init__(self, brand, model, year, battery_size):
        super().__init__(brand, model, year)
        self.battery_size = battery_size  # in kWh

    def electric_info(self):
        return f"{self.car_info()} | Battery: {self.battery_size} kWh"


# Example
ecar = ElectricCar("Tesla", "Model 3", 2024, 60)
print(ecar.electric_info())
```

```python
class Computer:
    def __init__(self, brand, processor, ram_gb):
        self.brand = brand
        self.processor = processor
        self.ram_gb = ram_gb

    def specs(self):
        return f"{self.brand} | {self.processor} | RAM: {self.ram_gb}GB"


class Laptop(Computer):
    def __init__(self, brand, processor, ram_gb, battery_life):
        super().__init__(brand, processor, ram_gb)
        self.battery_life = battery_life  # in hours

    def laptop_specs(self):
        return f"{self.specs()} | Battery: {self.battery_life} hrs"


# Example
lap = Laptop("Dell", "i5", 16, 8)
print(lap.laptop_specs())
```

```python
class ElectronicDevice:
    def __init__(self, brand, operating_system):
        self.brand = brand
        self.operating_system = operating_system

    def device_info(self):
        return f"{self.brand} | OS: {self.operating_system}"


class Smartphone(ElectronicDevice):
    def __init__(self, brand, operating_system, camera_mp):
        super().__init__(brand, operating_system)
        self.camera_mp = camera_mp

    def phone_info(self):
        return f"{self.device_info()} | Camera: {self.camera_mp}MP"


# Another child class (your choice): Tablet
class Tablet(ElectronicDevice):
    def __init__(self, brand, operating_system, screen_size_inch):
        super().__init__(brand, operating_system)
        self.screen_size_inch = screen_size_inch

    def tablet_info(self):
        return f"{self.device_info()} | Screen: {self.screen_size_inch}\""


# Example
ph = Smartphone("Apple", "iOS", 48)
tb = Tablet("Samsung", "Android", 11)
print(ph.phone_info())
print(tb.tablet_info())
```
