```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# 1. Send request to a website and extract top 5 headlines

url = "https://news.ycombinator.com/"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

headlines = soup.find_all("span", class_="titleline")

for headline in headlines[:5]:
    print(headline.get_text())
```

```python
# 2. Extract all hyperlinks text and URL from a webpage

url = "https://example.com"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

links = soup.find_all("a")

for link in links:
    text = link.get_text()
    href = link.get("href")
    print("Text:", text)
    print("URL:", href)
```

```python
# 3. Scrape and display titles of first 10 items using HTML tags

url = "https://news.ycombinator.com/"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

titles = soup.find_all("span", class_="titleline")

for title in titles[:10]:
    print(title.get_text())
```

```python
# 4. Extract text data with related information like author/name

url = "https://quotes.toscrape.com/"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

quotes = soup.find_all("div", class_="quote")

for quote in quotes:
    text = quote.find("span", class_="text").get_text()
    author = quote.find("small", class_="author").get_text()

    print("Quote:", text)
    print("Author:", author)
```

```python
# 5. Scrape top items from a website using headers User-Agent

url = "https://news.ycombinator.com/"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, "html.parser")

items = soup.find_all("span", class_="titleline")

for item in items[:5]:
    print(item.get_text())
```

```python
# 6. Extract data from an HTML table using table, tr, th, td tags

html = """
<table>
    <tr>
        <th>Name</th>
        <th>Age</th>
        <th>City</th>
    </tr>
    <tr>
        <td>Aman</td>
        <td>20</td>
        <td>Delhi</td>
    </tr>
    <tr>
        <td>Riya</td>
        <td>22</td>
        <td>Mumbai</td>
    </tr>
</table>
"""

soup = BeautifulSoup(html, "html.parser")

table = soup.find("table")
rows = table.find_all("tr")

for row in rows:
    columns = row.find_all(["th", "td"])
    data = [column.get_text() for column in columns]
    print(data)
```

```python
# 7. Scrape a table from a website and convert into pandas DataFrame

url = "https://www.w3schools.com/html/html_tables.asp"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

table = soup.find("table")

headers = []
for th in table.find_all("th"):
    headers.append(th.get_text())

rows = []
for tr in table.find_all("tr")[1:]:
    cells = tr.find_all("td")
    row = [cell.get_text() for cell in cells]
    rows.append(row)

df = pd.DataFrame(rows, columns=headers)

print(df)
```
