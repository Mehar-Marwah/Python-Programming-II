```python
#1
numbers = [10, 20, 30, 40, 50]

iterator = iter(numbers)

while True:
    try:
        print(next(iterator))
    except StopIteration:
        break
```

```python
# 2. Demonstrate what happens when next() is called more times than elements

fruits = ["apple", "banana", "mango"]

fruit_iterator = iter(fruits)

print(next(fruit_iterator))
print(next(fruit_iterator))
print(next(fruit_iterator))

# This will give StopIteration error because no element is left
# print(next(fruit_iterator))
```

```python
# 3. Iterate over an iterator using a for loop

colors = ["red", "blue", "green"]

color_iterator = iter(colors)

for color in color_iterator:
    print(color)
```

```python
# 4. Create an iterator using range() and print values step by step using next()

range_iterator = iter(range(1, 6))

print(next(range_iterator))
print(next(range_iterator))
print(next(range_iterator))
print(next(range_iterator))
print(next(range_iterator))
```

```python
# 5. Manually traverse a list using iter() and next() and handle StopIteration

students = ["Aman", "Riya", "Kavan", "Neha"]

student_iterator = iter(students)

while True:
    try:
        student = next(student_iterator)
        print(student)
    except StopIteration:
        print("No more elements left")
        break
```
