```
#DECORATOR
#A decorator is used to add extra behavior to a function without changing the original function.
#example :Your photo = original function
#Filter = decorator
#Photo stays same, but looks better
#Example
def my_decorator(func):
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper
def greet():
    print("Hello")

greet = my_decorator(greet)

greet()
```

```
#Q1
#a code that prints start before a function and ends afrter it
def my_decorator(func):
    def wrapper():
        print("Start")
        func()
        print("End")
    return wrapper

@my_decorator
def say_hello():
    print("Hello")

say_hello()
```

```
#Q2 Create a decorator that works with a function having one argument and prints a message before calling it.
def my_decorator(func):
    def wrapper(name):
        print("Calling function...")
        func(name)
    return wrapper

@my_decorator
def greet(name):
    print("Hello", name)

greet("Madiha")
```

```
# Q3 Write a decorator that doubles the result of a function.
def double_result(func):
    def wrapper():
        return func() * 2
    return wrapper

@double_result
def get_value():
    return 10

print(get_value())
```

```
#Q4 Create a decorator that counts how many times a function is called.
def count_calls(func):
    count = 0
    def wrapper():
        nonlocal count
        count += 1
        print("Called", count, "times")
        func()
    return wrapper

@count_calls
def say_hi():
    print("Hi")

say_hi()
say_hi()
say_hi()
```

```
#Q5 Create a decorator that checks if a number is even before calling the function.
def check_even(func):
    def wrapper(num):
        if num % 2 == 0:
            func(num)
        else:
            print("Number is odd, not allowed")
    return wrapper

@check_even
def show_number(num):
    print("Number is:", num)

show_number(4)
show_number(5)
```

```
#GENERATOR
#A generator is a special function that gives values one by one using yield.
#It does NOT give all values at once
# It gives values only when needed
#EXAMPLE
#Think of a water tap
#You open → water comes
#You close → it stops
# It does NOT pour the whole tank at once
#Q1 Write a generator that yields numbers from 1 to n.
def count(n):
  for i in range(1,n+1):
    yield i
for i in count(5):
  print(i)
```

```
#Q2 Create a generator that yields square of numbers from 1 to n.
def square(n):
  for i in range(1 ,n+1):
    yield i*i #Take the value of i, square it, and give it (return it) one by one
for i in square(5):
  print(i)
```

```
#Q3 Write a generator that yields even numbers up to n.
def even(n):
  for i in range (1, n+1):
    if i % 2==0:
       yield i
for i in even(10):
  print(i)
```

```
#Q4 generator to calculat sum of first n numbers
def numbers(n):
    for i in range(1, n+1):
        yield i

total = 0

for i in numbers(5):
    total += i

print("Sum =", total)
```

```
#Q5 odd numbers
def odd(n):
    for i in range(1, n+1):
        if i % 2 != 0:
            yield i

for i in odd(10):
    print(i)
```
