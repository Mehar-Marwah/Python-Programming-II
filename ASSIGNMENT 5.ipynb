```
import time

# 1. Decorator: print before and after function execution
def before_after(func):
    def wrapper():
        print("Before function execution")
        func()
        print("After function execution")
    return wrapper

@before_after
def greet():
    print("Hello!")

greet()
```

```



# 2. Decorator: custom message before function call
def custom_message(func):
    def wrapper():
        print("This is a custom message before the function")
        func()
    return wrapper

@custom_message
def say_hello():
    print("Hello from original function")

say_hello()
```

```

# 3. Decorator: log function name and two parameters
def log_two_args(func):
    def wrapper(a, b):
        print("Function name:", func.__name__)
        print("Parameters:", a, b)
        return func(a, b)
    return wrapper

@log_two_args
def add(a, b):
    return a + b

print(add(10, 20))
```

```

# 4. Decorator: measure execution time of no-argument function
def measure_time(func):
    def wrapper():
        start = time.time()
        result = func()
        end = time.time()
        print("Execution time:", end - start, "seconds")
        return result
    return wrapper

@measure_time
def slow_function():
    time.sleep(1)
    print("Function completed")

slow_function()
```

```
# 5. Decorator: check if user is admin
def admin_required(func):
    def wrapper(user):
        if user == "admin":
            return func(user)
        else:
            print("Access denied")
    return wrapper

@admin_required
def dashboard(user):
    print("Welcome to admin dashboard")

dashboard("admin")
dashboard("guest")
```

```
# 6. Decorator: cache results of function with one integer argument
def cache_result(func):
    cache = {}

    def wrapper(n):
        if n in cache:
            print("Returning cached result")
            return cache[n]
        result = func(n)
        cache[n] = result
        return result

    return wrapper

@cache_result
def square(n):
    print("Calculating result")
    return n * n

print(square(5))
print(square(5))
```

```
# 7. Decorator: works with *args and **kwargs
def print_before_after(func):
    def wrapper(*args, **kwargs):
        print("Before execution")
        result = func(*args, **kwargs)
        print("After execution")
        return result
    return wrapper

@print_before_after
def student_info(name, age):
    print("Name:", name)
    print("Age:", age)

student_info("Kavan", age=18)
```

```

# 8. Decorator: double the return value
def double_result(func):
    def wrapper():
        result = func()
        return result * 2
    return wrapper

@double_result
def get_number():
    return 10

print(get_number())
```
